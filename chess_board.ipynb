```python
import matplotlib.pyplot as plt
import numpy as np

class Chess_Board: 

    white  = [1, 1, 1]
    blue = [0, 1, 1]
    red = [1, .2, 0]


    def __init__ (self):

        self.grid = np.zeros([8,8,3])
        self.place_white()
        self.blue_x = 0
        self.blue_y = 0
        self.red_x = 0
        self.red_y = 0


    def place_white(self):

        for x in range(8):
            for y in range(8):
                if (x% 2 == 0 and y % 2 == 0) or (x % 2 == 1 and y % 2 == 1):
                    self.grid[x,y] = ChessBoard.white

    def draw_board(self):

        plt.imshow(self.grid)


    def add_blue(self, row, col):

        self.grid[row, col]=Chess_Board.blue
        Chess_Board.blue_x = row
        Chess_Board.blue_y = col
        self.draw_board()

    def add_red(self, row, col):

        self.grid[row, col]=Chess_Board.red
        Chess_Board.red_x = row
        Chess_Board.red_y = col
        self.draw_board()

    
    def diagonal (self):
        
        sub_x = abs(Chess_Board.red_x - Chess_Board.blue_x)
        sub_y = abs(Chess_Board.red_y - Chess_Board.blue_y)
        return sub_x == sub_y


    def is_under_attack (self):

        diagonal = self.is_diagonal()
        
        if self.red_x == self.blue_x:
            return True
        elif self.red_y == self.blue_y:
            return True
        elif diagonal:
            return True
        else:
            return False  
    

board = ChessBoard()
board.add_red(6,4)
board.add_blue(3,0)
print(board.is_under_attack())
board.render()


board.add_red(3,4)
board.add_blue(3,0)
assert board.is_under_attack() == True


board.add_red(2,4)
board.add_blue(6,4)
assert board.is_under_attack() == True

board.add_red(7,4)
board.add_blue(3,0)
assert board.is_under_attack() == True

board.add_red(6,4)
board.add_blue(3,0)
assert board.is_under_attack() == False


print('Total success woooohooooo!! ')
```
